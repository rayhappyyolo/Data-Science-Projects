```python
# Check for missing values
missing_values = df.isnull().sum()

# Summary statistics
summary_statistics = df.describe()

missing_values, summary_statistics
```

```python
df['comments'] = pd.to_numeric(df['comments'], errors='coerce')
df['likes'] = pd.to_numeric(df['likes'], errors='coerce')
features = df[['comments', 'likes']].dropna()
```

```python
#Clustering Analysis
#Applying the Elbow Method to Determine Optimal Clusters:

from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

sample_data = features.sample(frac=0.1, random_state=42)
wcss_sample = []
for i in range(1, 11):
    kmeans_sample = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans_sample.fit(sample_data)
    wcss_sample.append(kmeans_sample.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss_sample)
plt.title('Elbow Method (Sampled Data)')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS (Sampled)')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
#K-Means Clustering with 4 Clusters:

kmeans_final = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=42)
cluster_labels = kmeans_final.fit_predict(sample_data)
sample_data['Cluster'] = cluster_labels
```

```python
#Analysis of Clusters
#Cluster Summary:

cluster_summary = sample_data.groupby('Cluster').agg({
    'comments': ['mean', 'min', 'max'],
    'likes': ['mean', 'min', 'max'],
    'Cluster': 'count'
}).rename(columns={'Cluster': 'Count', 'comments': 'Comments', 'likes': 'Likes'})
```

We then proceeded to analyze these clusters, providing insights into the engagement levels of posts within each cluster. The summary statistics and distribution analyses helped us understand the characteristics of each cluster:

Cluster 0 (General Engagement): Posts with moderate engagement levels, representing the majority of the data.

Cluster 1 (High Likes, High Comments): Posts with extremely high engagement, indicating viral or highly popular content.

Cluster 2 (Moderate to High Engagement): Posts with notably high engagement but not as high as Cluster 1.

Cluster 3 (Top Engagement): The highest engagement posts, likely representing top-tier viral content.

```python
#Centroid Analysis:

centroids = kmeans_final.cluster_centers_
centroids_df = pd.DataFrame(centroids, columns=['Average Comments', 'Average Likes'])
centroids_df['Cluster'] = centroids_df.index
```

```python
#Distribution Analysis:

import seaborn as sns

plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Cluster', y='comments', data=sample_data)
plt.title('Distribution of Comments by Cluster')
plt.subplot(1, 2, 2)
sns.boxplot(x='Cluster', y='likes', data=sample_data)
plt.title('Distribution of Likes by Cluster')
plt.tight_layout()
plt.show()
```
